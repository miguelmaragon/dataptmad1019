## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd 
data = pd.read_csv('../../lab-data_cleaning/weather-raw.csv')
# Preview the first 5 lines of the loaded data 
data
```

```python
data.describe()
```

```python
null_cols = data.isnull().sum()
null_cols
```

```python
#data[['d1','d31'] = data[['d1','d31']].fillna(0)
#data.loc[:,'d1':'d31'] = data.loc[:,'d1':'d31'].fillna(0)
```

```python
null_cols = null_cols[null_cols > 0] / len(data) * 100
null_cols
```

```python
threshold = 95.00

#null_cols = null_cols[null_cols > 0] / len(data) * 100


null_filter = null_cols > threshold

drop_cols = list(null_cols[null_filter].index)

drop_cols
```

```python
data.drop(drop_cols, axis=1, inplace=True)

len(data.columns)
```

```python
data.loc[:,'d1':'d31'] = data.loc[:,'d1':'d31'].fillna(0)
data.T
```
